## Analysis Question 1 - What product was purchased most?
To answer this question, we use the groupby method to subset the dataframe by the columns - 'Description', 'Size' and 'Sales_Qty' to tell us which product was sold the most by the chain.

```python
# MOST  PURCHASED PRODUCT.
# subset the dataset grouping by the description and size, subsetting sales quantity and aggreagting the minimum value, 
# maximum value and the sum of all the sales on each item

subset_for_highest_sales = df.groupby(['Description','Size'])['Sales_Qty'].agg([min, max, sum])
subset_for_highest_sales # make a call to the subset

# sort the subset dataframe in descending order to find out which product has the highest sum(highest sales)
sorted_subset = subset_for_highest_sales.sort_values('sum', ascending=False)
# sorted_subset
```

With PANDAS, we can visualize our data using bar chart

```python
subset_for_highest_sales = df.groupby(['Description','Size'])['Sales_Qty'].agg([min, max, sum])
sorted_subset = subset_for_highest_sales.sort_values('sum', ascending=False).head(20)
sorted_subset.plot.bar()
```

### INSIGHT
From the subsetted dataframe, we can see that both Smirnoff 80 Proof-50mL and Yukon Jack-50mL has the same max sales quantity but Smirnoff 80 Proof-50mL has the highest total sales quantity. Of all the different products, Smirnoff 80 Proof-50mL was sold the most.

## Analysis Question 2 - What bottle size is most sold and the least sold?
To answer this question, we will subset the data frame by columns 'Size' and 'Sales_Qty' to tell us which bottle size is most preferred by customers of the chain and which one is least preferred.

```python
# MOST SOLD  AND LEAST SOLD SIZE.
# subset the dataset grouping by the size, subsetting sales quantity and aggreagting the minimum value, 
# maximum value and the sum of all the sales on each item

subset_for_highest_size = df.groupby('Size')['Sales_Qty'].agg([min, max, sum])

# sort the subset dataframe in descending order to find out which size has the highest sum(highest sales) and lowest.
sorted_subset2 = subset_for_highest_sales.sort_values('sum', ascending=False)
sorted_subset2
```

## Analysis Question 3 - How does price affect the sales of a product?
To answer this question, we will subset dataframe by the columns 'Unit_price($) and 'Sales_Qty' to tell us how different product prices influence the buying patterns of customers

```python
# DOES PRICE HAVE IMPACT ON THE SALES QUANTITY OF AN ITEM?

# Firsly subset the dataframe by grouping by price and subsetting sales quantity for easy comparison and aaggreagte it
subset_for_price = df.groupby('Unit_price($)')['Sales_Qty'].agg([min, max, sum])

# display the sorted version of the subsetted dataframe in both ascending and descending order
print(subset_for_price.sort_values('sum', ascending= False).head(10))
print(subset_for_price.sort_values('sum', ascending= True).head(10))
```

## Analysis Question 4 - Which store sold the most?
To answer this question, we will subset dataframe by the columns 'Store' and 'Sales_Qty' to tell us which of the chain's stores raked up the most sales

## Analysis Question 6 - Which Vendor has the highest variety of products sold?
To answer this question, we use the groupby method to subset the dataframe by the columns - 'VendorName', 'VendorNo' and 'Brand' to tell us which Vendor produces the most products sold throughout the chain.

```python
# WHICH VENDOR HAS THE HIGHEST VARIETY OF PRODUCTS SOLD

# subset the dataset grouping by the Vendor name and vendor number, subsetting sales quantity and using the
# nunique() method to get the number of unique varieties.
subset_for_highest_Variety = df.groupby(['VendorName', 'VendorNo'])['Brand'].nunique().sort_values(ascending=False)
subset_for_highest_Variety
```

### INSIGHT

From the result of the analysis, we can deduce that the Vendor that has the highest variety of products sold is MARTIGNETTI COMPANIES with a totla of 1,042 products.
