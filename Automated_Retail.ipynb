```python
#define function calculating total sales
def calculate_total_sales(morning_shift_sales,evening_shift_sales):
    """
    Calculates the total sales for both shifts.
"""
    
    total_sales = morning_shift_sales + evening_shift_sales
    return total_sales
    
```

```python
# defining the function calculating workers salaries
def workers_salary(hourly_rate,hours_worked):
    """
    Calculates the worker's salary based on hourly rate and hours worked.
    """
# summing the parameters
    salary = hourly_rate * hours_worked
    return salary
```

```python
# calculating the the profit of the enterprize
# defining the function
def profit_loss(total_sales,total_cost):
    """
    Calculates the profit for the day by subtracting total cost from total sales.
    """
    profit = total_sales - total_cost
    return profit
```

```python
# defining the Function to calculate tips for each shift
def tips_per_shift(sales_per_shift):
    return [sales * 0.02 for sales in sales_per_shift]  # Calculate tips for each shift
```

```python
# defining the function calculating tips per day

def total_tips(morning_tips,evening_tips):
    """
    Calculates the tips for a single shift based on a 2% tip rate.
    """
    total = morning_tips + evening_tips
    return total
```

```python
def retail_business_calculation():
    """
    This function allows users to choose and perform various retail business calculations.
    """
    morning_shift_sales = []
    evening_shift_sales = []
    total_daily_cost = []
    calculated_tips = {}  # Store calculated tips


    while True:
        user_input1 = input("Enter morning sales (or 'q' to quit): ")
        user_input2 = input("Enter evening sales (or 'q ' to quit): ")
        user_input3 = input("Enter total daily expenses (or 'q' to quit): ")
        

        if user_input1.lower() == "q" or user_input2.lower() == "q" or user_input3.lower() == "q":
            break

        try:
            morning_sales = float(user_input1)
            evening_sales = float(user_input2)
            total_spent = float(user_input3)
            morning_shift_sales.append(morning_sales)
            evening_shift_sales.append(evening_sales)
            total_daily_cost.append(total_spent)
        except ValueError:
            print("Invalid input, please enter a number.")

    print(f"\nYour list of morning sales is: {morning_shift_sales}")
    print(f"Your list of evening sales is: {evening_shift_sales}")
    print(f"Your list of total expenses is: {total_daily_cost}")

    print(f"Sum of morning sales is: {sum(morning_shift_sales):.1f}")
    print(f"Sum of evening sales is: {sum(evening_shift_sales):.1f}")
    print(f"Sum of total expenses is: {sum(total_daily_cost):.1f}")
    
    # Creating Retail business operations menu and user choices

    while True:
        print("\n Retail Business Automation\n ")
        print("1. Calculate Total Sales for the Day")
        print("2. Calculate Worker's Salary")
        print("3. Calculate Profit")
        print("4. Calculate tips for a shift")
        print("5. Calculate Total Tips for the Day")
        print("6. Exit")

        try:
            choice = int(input("Enter your choice (1-6): "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        
        if choice == 1:
            total_sales = calculate_total_sales(sum(morning_shift_sales), sum(evening_shift_sales))
            print(f"Total Sales for the Day: €{total_sales:.1f}")

        elif choice == 2:
            hourly_rate = float(input("Enter hourly rate: €"))
            hours_worked = float(input("Enter hours worked: "))
            worker_salary = workers_salary(hourly_rate, hours_worked)
            print(f"Worker's Salary: €{worker_salary:.1f}")

        elif choice == 3:
            total_sales = sum(morning_shift_sales) + sum(evening_shift_sales)
            total_cost = sum(total_daily_cost)
            profit_los = profit_loss(total_sales, total_cost)
            print(f"Profit: €{profit_los:.1f}")

        elif choice == 4:
            if not calculated_tips:  # Calculate tips only if not already done
                morning_tips = tips_per_shift(morning_shift_sales)
                evening_tips = tips_per_shift(evening_shift_sales)
                calculated_tips = {"morning": morning_tips, "evening": evening_tips}
            print(f"Tips for the morning shift: €{sum(calculated_tips['morning']):.1f}")
            print(f"Tips for the evening shift: €{sum(calculated_tips['evening']):.1f}")

        # Option 5: Calculate total tips for the day
        elif choice == 5:
            if calculated_tips:
                total_tip = sum(calculated_tips["morning"]) + sum(calculated_tips["evening"])
                print(f"Total Tips for the Day: €{total_tip:.1f}")
            else:
                print("Please calculate tips for each shift first.")

        elif choice == 6:
            print("Exiting . Thank you!")
            break

        else:
            print("Choice not valid. Pls enter a number between 1 and 6.")
```

```python
retail_business_calculation()
```
